# تحلیل استراتژی سانتر در فیفا

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### خواندن داده ها

```python
fifa11 = pd.read_csv('../data/fifa/fifa11.csv')
fifa21 = pd.read_csv('../data/fifa/fifa21.csv')
fifa21.head()
```

```python
height = pd.concat([pd.DataFrame({'fifa11': fifa11['height']}), pd.DataFrame({'fifa21': fifa21['height']})], axis=1)
height.head()
```

### نمودار هیستوگرام و چگالی برای قد ها

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
fig.suptitle('FIFA11 vs FIFA21 Heights', fontsize='xx-large')

height.plot(kind='hist', ax=ax[0], alpha=0.5, density=True)
height.plot(kind='kde', ax=ax[1], alpha=0.5)

plt.show()
```
