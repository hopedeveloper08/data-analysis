# تحلیل عناصر و عیار از داده های شرکت یونیدرو

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### خواندن داده ها

```python
df = pd.read_csv('../data/unidro_data.csv')
df.head()
```

### ماتریس همبستگی

```python
corr_matrix = df.corr()
corr_matrix.head()
```

# هیت مپ

```python
mask = np.tril(np.ones((31, 31)))
```

```python
sns.set_theme()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, cmap='jet', mask=mask)
plt.show()
```
